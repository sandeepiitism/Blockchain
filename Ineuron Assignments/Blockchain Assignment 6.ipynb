- Updating state variables
- Reading state variables
- Logic execution

### 4. What is the difference between storage variable and memory variable with respect to ethereum blockchain?

<b> Storage</b>: This is global memory available to all functions within the contract. This storage is a permanent storage that
Ethereum stores on every node within its environment. <br>
<b> Memory</b>: This is local memory available to every function within a contract. This is short lived and fleeting memory that gets torn down when the function completes its execution.
