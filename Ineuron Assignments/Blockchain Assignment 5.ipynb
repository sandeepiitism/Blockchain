### 1. What is gwei? What is the relationship between wei, gwei and ether?

<b>Gwei</b> is a unit of Ether to measure the cost of what is called “gas”. Gwei is the most commonly used denomination of Ether. <br>
`1,000,000,000 Wei (10^9) is 1 Gwei` <br>
`1 Ether = 1,000,000,000 Gwei`

<b>Wei</b> – Wei is the smallest (base) unit of Ether. It’s like a Cent to USD and a Satoshi to BTC. It takes a great number of Wei to make an ETH; 10^18 wei, to be exact. It’s like a digital dust.<br>
`1 Ether = 1,000,000,000,000,000,000 Wei`<br>

<b>Ether (ETH)</b> is the cryptocurrency generated by the Ethereum protocol as a reward to miners in a proof-of-work system for adding blocks to the blockchain.




### 2. What are the message parameters of a smart contract call?

Contracts can call other contracts or send Ether to non-contract accounts by the means of message calls. Message calls are similar to transactions, in that they have a source, a target, data payload, Ether, gas and return data.

 `selfdestruct`, `delegatecall` or `callcode` are the message parameters of a smart contract call.

### 4. What is the difference between gas price, gas limit and gas units?

In smart contracts there is a price of executing each line of code. Possible operations the EVM can perform are hard-coded in the Ethereum protocol and the lines of code result into <b> gas price </b> which will be x amount depending on the gaas charges of opcodes used whereas gas limit is a limit after which contract will stop executing.

We use <b> gas limit</b> so that if anyhow our contract goes in infinite then it will stop after gas limit reached so that it wont cost us too many ethers.

<b>Gas unit </b> is the total cost of an Ethereum transaction is actually the amount of necessary gas multiplied by the price in GWei per gas unit.
