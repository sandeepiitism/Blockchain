### 1. What is the difference between byte code and abi. Show the byte code and abi of any compiled solidity smart contract of your choice. Show the bytecode to opcode transformation through : https://etherscan.io/opcode-tool.

<b>Application Binary Interface (ABI)</b> as an interface consisting of all external and public function declarations along with their parameters and return types. The ABI defines the contract and any caller wanting to invoke any contract function can use the ABI to do so. <br> <br>
The <b>bytecode </b>is what represents the contract and it is deployed in the Ethereum ecosystem. The bytecode is required during deployment and ABI is needed for invoking functions in a contract.
