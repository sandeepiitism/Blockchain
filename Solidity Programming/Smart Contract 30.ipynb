## Fallback and Receive

When <b>Fallback</b> is used for Ethers

When <b>Fallback</b> is used for Data

When <b>Receive</b> is used

When <b>fallback</b> and <b>receive</b> both are used and only Ether is transacted ---> Priority goes to receive

When <b>fallback</b> and <b>receive</b> both are used and both Data and Ether is transacted ---> Priority goes to Fallback
