## Data Types

Uint---> unsigned integer starts from 0, positive Integers
- uint256: 0 to 2^256 - 1
- uint8: 0 to 2^8 - 1

Int ---> signed Integer includes negative numbers
- int256: -2^255 to +2^255-1
- int8: -2^7 to +2^7-1

Bool ---> true or False <br>

Address ---> 20bits hexadecimal, account address

Bytes32 ---> any hexadecimal numbers that can be converted to 0's and 1's.

`note`: 
- Lesser the int or uint lesser the gas fee
- use bytes32 in place of string if you have a limited text/words, if not use string.
