## Hash Collision

 encode vs encodePacked

<b>encode</b>: used to avoid hash collision <br>
    
<b>encodePacked</b>: can create hash collision even used with keccak256    

encode is used with <b> abi.encode</b> & endepac used with <b>abi.encodePacked</b> which compresses the abi.encode

But <b>abi.encodePacked</b> while compressing has ait has a disadvantage of having hash collision

"abc", "efg" as input

"ab", "cdef" as input but hash doesn't changes ---> Hash collision
